```python
# -*- coding: utf-8 -*-
"""
Created on Mon Feb 29 11:17:21 2016

Author: Richard Krantz

(Python 3.5)

Description: Transmission coefficient for a delta function  structure
as a function of the reduced input energy 
"""
# Template imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
ax = plt.gca()
# import scipy as sci
# from scipy.special import factorial as fac
# import sympy as sym
# ----------------------------------------------------------------------
plt.close('all')
# ----------------------------------------------------------------------
vo=10;# "Strength of the delta function potential
for e in np.linspace(0.01,250,5001):
    b=vo/(2*np.sqrt(e));
    k=np.sqrt(e)
    T1=1/(1+(vo/(2*k))**2)
    T2=1/((1+b**2)**2 + b**4 -2*(b**2)*(1+b**2)*(np.cos(2*k)))
    plt.figure(1); # Format Figure #1
    plt.axis('tight')  # Minimize space around plot
    plt.title("Transmission for Delta Function(s)", size=20)
    plt.xlabel('Reduced Input Energy', size=15)
    plt.ylabel('Transmission Coefficient', size=15)
    plt.grid(b=True, which='major', color='b',linestyle='-')
    plt.plot(e,T1,'b  .');plt.plot(e,T2,'r  .')
# Blue line is for a single delta function
# Red line is for 2 delta functions - resonant tunneling structure
```
